```python
file_path = "mapping/us_states/cb_2018_us_state_500k.shp"
states = gpd.read_file(file_path)
type(states)
```

```python
#Make the map look a little familiar by reprojecting its coordinates to Mercator
states = states.to_crs("EPSG:3395")
```

```python
states.plot()
```

```python
hires_map = states.boundary.plot(figsize=(50, 28),  color="Gray", linewidth=4)
```

```python
df = pd.read_csv("mapping/counties_latlng.csv")
df.head(2)
```

```python
def make_point(row):
    return Point(row.Lat_H, row.Lng_H)

#Go thru every row, make a point out of its latitude and longitude
points_H = df.apply(make_point, axis=1)

#Make a new GeoDF using the data from old df, but add in the geometry
counties_H = gpd.GeoDataFrame(df, geometry=points_H)

#A CSV doesn't come with a CRS, therfore must use a standard shape of earth
counties_H.crs = {'init': 'epsg:4326'}

counties_H.head()
```

```python
ax = states.plot(figsize=(100,90))
counties_H.plot(markersize=25, color='black', alpha=0.5, ax=ax)
ax.axis('off')
```
