Data: CSV file with ALL state data updated to real time. 

```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Store filepath in a variable
file_one = "CSV data/all-states-history.csv"
state_df = pd.read_csv(file_one)
```

```python
#Show the header of df
state_df.head()
```

```python
#Check if state_df['date'] is a Series and check dtype
state_df['date']
```

```python
state_df['date'] = pd.to_datetime(state_df['date'])
state_df['date']
```

```python
start_date = state_df[state_df["date"].dt.date.astype(str) == '2020-03-01'] 
start_date
```

```python
end_date = state_df[state_df["date"].dt.date.astype(str) == '2020-12-31'] 
end_date
```

```python
start_end = pd.merge(start_date, end_date, on=("date"), how="left")
```
