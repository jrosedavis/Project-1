```python
# Dependencies
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
# file_one = "data/us-states.txt"
# us_states = pd.read_csv(file_one)
file_two = "data/us-counties.txt"
us_counties = pd.read_csv(file_two)
```

```python
us_counties
```

```python
filter_VT = us_counties.loc[us_counties['state'] == 'Ohio', :]
#filter_VT
```

```python
grouped = filter_VT.groupby(['county'])
```

```python
maxVT_cases = grouped['cases'].max()
maxVT_cases
VT_counties = pd.DataFrame(maxVT_cases)
VT_counties
```

```python
VT_counties = VT_counties.sort_values(['cases'],ascending='False')
VT_counties
```
