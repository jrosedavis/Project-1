```python
# Dependencies
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
file_one = "data/us-states.txt"
us_states = pd.read_csv(file_one)
file_two = "data/us-counties.txt"
us_counts = pd.read_csv(file_two)
```

```python
us_states
```

```python
# clean_states_df = us_states.dropna(how="any")
# clean_states_df
```

```python
grouped = us_states.groupby(['state'])#.apply(', '.join).reset_index()
```

```python
max_cases = grouped['cases'].max()
```

```python
remove_terr = pd.DataFrame(max_cases) 
fifty_states = remove_terr.drop(['Guam', 'Puerto Rico', 'Northern Mariana Islands', 'Virgin Islands', 'District of Columbia'])
```

```python
fifty_states = fifty_states.sort_values(['cases'],ascending='False')
```

```python
#Create a bar chart based off of the the group series from count_DR
panda_chart = fifty_states.plot(kind='barh', figsize=(15, 10))
panda_chart.set_xlabel("Total Number")
panda_chart.set_ylabel("States")
plt.title("States and Maximum Cases")
# plt.tight_layout()
plt.show()
```
